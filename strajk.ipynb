## **–ü—Ä–µ–≤–∑–µ–º–∞—ö–µ –∏ —á–∏—Å—Ç–µ—ö–µ –Ω–∞ –ø–æ–¥–∞—Ç–æ—Ü–∏**

```python
import tweepy as tw

consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''

auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth)
```

```python
search_terms = ('—à—Ç—Ä–∞—ò–∫ -filter:tweets OR -filter:retweets OR filter:replies')
query = tw.Cursor(api.search, q=search_terms, include_entities=True).items(100)
tweets = [tweet.text for tweet in query]
print(tweets)
```

```python
import re
tvitovi = []

for tweet in tweets:
  tweet = re.sub(r"http\S+|www\S+|https\S+", '', tweet, flags=re.MULTILINE)
  tweet = re.sub(r'RT @:?(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+','', tweet)
  tweet = re.sub(r'@:?(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+','', tweet)
  #print(tweet)
  tvitovi.append(tweet)

print(tvitovi)
```

## **–ù–∞—ò—á–µ—Å—Ç–∏ –ø–æ—ò–∞–≤—É–≤–∞—ö–∞ –Ω–∞ –∫–æ—Ä–∏—Å—Ç–µ–Ω–∏ –∑–±–æ—Ä–æ–≤–∏**

–ü–æ–¥–µ–ª–µ–Ω–∏ –≤–æ 4 –≥—Ä—É–ø–∏:
1. –≥–ª–∞–≥–æ–ª–∏
2. –∏–º–µ–Ω–∫–∏
3. –ø—Ä–∏–¥–∞–≤–∫–∏
4. –ø—Ä–∏–ª–æ–∑–∏

```python
import spacy
import re
import itertools
import numpy as np
from spacy import displacy


def top_pos(doc, pos, n):
  

    pos_count = {}
    for token in doc:
        # ignore stop words
        if token.is_stop:
            continue

        if token.pos_ == pos:
            if token.lemma_ in pos_count:
                pos_count[token.lemma_] += 1
            else:
                pos_count[token.lemma_] = 1

    # sort by values, but before get only those keys where value > 1;
    # I want lemmas that appear more than one
    # lastly, get the first n results
    result = sorted({k: v for (k, v) in pos_count.items() if v > 1}.items(),
                    key=lambda kv: kv[1], reverse=True)[:n]

    print("top {} {} {}".format(n, pos, result))


def top_entities(doc, n):
    """Finds the top n spaCy entities
    Parameters:
    doc: spaCy's doc
    n: how many entities
    """
    
    entities = {}
    # named entities
    for ent in doc.ents:
        # Print the entity text and its label
        if ent.text in entities:
            entities[ent.text] += 1
        else:
            entities[ent.text] = 1
    result = sorted(entities.items(), key=lambda kv: kv[1], reverse=True)[:n]
    print("top {} entities {}".format(n, result))



if __name__ == "__main__":
  string_tvitovi = ' '.join(str(item) for item in tvitovi)  
  doc = nlp(string_tvitovi)

  top_pos(doc, 'VERB', 20)
  top_pos(doc, 'NOUN', 20)
  top_pos(doc, 'ADJ', 20)
  top_pos(doc, 'ADV', 20)
  top_entities(doc, 30)
    
```
